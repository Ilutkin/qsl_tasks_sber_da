```python
import requests
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
```

## Список объектов

```python
offset_ = 2001
limit_ = 1000
url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
res = requests.get(url)
objects_data = res.json()
```

```python
objects_data.keys()
```

```python
# Собрать информацию о всех строящихся объектах на сайте "наш.дом.рф"
objects_list = pd.json_normalize(objects_data.get('data').get('list'))
# objects_list.set_index('objId', inplace=True)
objects_list
```

```python
obj_ID = list(objects_list['objId'])
obj_ID
```

# Данные по определенному объекту

```python
url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/25002'
```

```python
obj_description = pd.DataFrame()
description_obj_list = []
for ids in tqdm(obj_ID):
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{ids}'
    res = requests.get(url)
    v = res.json()
    description_obj_list.append(v)
obj_description = pd.json_normalize(description_obj_list)
obj_description.tail()
```

```python
# Cохранить ее в pandas dataframe
all_obj_info = pd.merge(objects_list,obj_description, left_on=['objId'], right_on=['data.id'], left_index=False, right_index=False, how='left').copy()
all_obj_info.head(5)
```

```python
# Cохранить pandas dataframe в excel
all_obj_info.to_excel('DMRF.xlsx', engine='xlsxwriter')  
```

```python
# Cохранить pandas dataframe в pickle
pd.to_pickle(all_obj_info, "./all_obj_info.pkl")  
```

```python
# Задание 2
# Выберите регион
all_obj_52_reg = all_obj_info[all_obj_info['rpdRegionCd'] == 52].copy()
all_obj_52_reg.head(3)
```

```python
all_obj_52_reg['city_name'] = all_obj_52_reg['objAddr'].str.extract(r'\bг\s*([^,]+)')
all_obj_52_reg.head(3)
```

```python
# Сравните среднюю цену квадратного метра в городах этого региона
comparison = pd.pivot_table(all_obj_52_reg, index=["city_name"], values=["data.objPriceAvg"],  aggfunc=[np.mean, len])
comparison
```

```python
# Определите количество строящихся объектов и квартир в регионах
all_obj_reg = pd.pivot_table(all_obj_info, 
                             index=['data.developer.regRegionDesc'], 
                             values=['objElemLivingCnt', 'objSquareLiving', 'objId'],  
                             aggfunc={'objElemLivingCnt': np.sum, 'objSquareLiving': np.sum, 'objId': len}, 
                             fill_value=0)
all_obj_reg.sort_values(by='objSquareLiving', ascending = False).head(30)
```
